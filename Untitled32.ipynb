```python
class car:
  def __init__(self, model, color, seater, horsepower):
    self.model = model
    self.color = color
    self.seater = seater
    self.horsepower = horsepower

car1 = car("Macedes", "Black", "four_seater", 2200)

print(car1.model)
print(car1.color)
print(car1.seater)
print(car1.horsepower)
```

```python
print(f"Car1 is {car1.model}, {car1.color} in color and can carry up to {car1.seater} with {car1.horsepower} horsepower.")
```
