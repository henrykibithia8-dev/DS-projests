# Simple linear regression

## Import the relevant libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv("Simple linear regression.csv")
data
```

## Create the regression

### Declare the dependent and independent variables

```python
x = data.drop(['GPA'], axis= 1)
y = data['GPA']
```

```python
x.shape
```

```python
y.shape
```

### Regression itself

```python
reg = LinearRegression()
```

```python
reg.fit(x,y)
```

```python
#lets get R_squared
reg.score(x, y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
import numpy as np
#predict new data
reg.predict(np.array(1740).reshape(-1, 1))
```

```python
import numpy as np
new_data = pd.DataFrame(data=[[1740, np.nan], [1760, np.nan]], columns=["SAT", "GPA"])
new_data
```

```python
reg.predict(new_data[['SAT']])
```

```python
plt.scatter(x,y)
yhat = reg.coef_*x + reg.intercept_
fig = plt.plot(x, yhat, c = "w")
plt.show()
```
