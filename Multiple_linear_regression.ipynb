```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

from sklearn.linear_model import LinearRegression
```

```python
#load data
data = pd.read_csv("https://github.com/henrykibithia8-dev/DS-projests/raw/main/Multiple%20linear%20regression.csv")

data
```

```python
data.describe()
```

Create Multiple LinearRegression

```python
#declaring the dependent and independent variable
x = data.drop(["GPA"], axis = 1)
y = data["GPA"]
```

Declare the LinearRegression

```python
#model
reg = LinearRegression()
```

```python
#fit the regression into data
reg.fit(x,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.score(x,y)
```

```python
reg.predict(pd.DataFrame(data=[[5000, 2]], columns=["SAT", "Rand 1,2,3"]))
```

```python
reg.score(x,y)
```

```python
x.shape
```

```python
r2 = reg.score(x,y)
n = x.shape[0]
p = x.shape[1]

adr2 = 1-(1-r2)*(n-1)/(n-p-1)
adr2
```

Feature selection

```python
from sklearn.feature_selection import f_regression

print(f"f_statistics: {f_regression(x,y)}")
```

```python
p_values = f_regression(x,y)[1]
p_values.round(3)
```

Creating a summary table

```python
#dataframe
df = pd.DataFrame(data=["SAT", "Rand 1 2 3"], columns=["features"])
df
```

```python
reg_summary= pd.DataFrame(data = x.columns.values, columns= ["Features"])
df
```

```python
#now  summary for coef and p_values
reg_summary["coeficient"] = reg.coef_
reg_summary["p_values"] = p_values.round(3)
reg_summary
```
