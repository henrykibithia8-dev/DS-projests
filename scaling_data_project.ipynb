```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/henrykibithia8-dev/raw-files/main/Multiple%20linear%20regression.csv")
data
```

```python
X = data.drop(["GPA"], axis = 1)
y = data["GPA"]
```

```python
X
```

```python
#import standard scaler
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

```

```python
#fit scaler in our x data
scaler.fit(X)
```

```python
#transform our data
x_scaled = scaler.transform(X)
x_scaled
```

```python
#fit  a linear model using scaled values
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x_scaled,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
#creaing a summary
summary = pd.DataFrame([["intercept"], ["SAT"], ["Rand 1 2 3"]], columns = ["Features"])
summary["weight"] = reg.intercept_, reg.coef_[0],reg.coef_[1]
summary
```

```python
#making predictions
n_data = pd.DataFrame([[1700,2], [1800,1]], columns=["SAT", "Rand 1,2,3"])
n_data
```

```python
reg.predict(n_data)
```

```python
#the above output values are bigger because they are not scaled
#now lets standardize the new data using our scals
sn_data = scaler.transform(n_data)
reg.predict(sn_data)
```

```python
snn_data = pd.DataFrame([[1700], [1800]], columns=["SAT"])
sn_data = scaler.transform(snn_data)
```
